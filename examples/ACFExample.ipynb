# Using the autocorrelation function to find a periodic signal


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# use seaborn for matplotlib settings
import seaborn; seaborn.set()
```

## Standard ACF
The first exmple data we will use will be Q1-Q15 photometry of KOI-63. We're using Kepler data as the example because the standard ACF requires evenly-sampled data.  This data is pre-conditioned to have no gaps (gaps have been set to zero).

```python
data = np.load('koi63.npy')
t, y = (data[:,0], data[:,1])

fig, ax = plt.subplots()
ax.plot(t, y, 'o', ms=0.5);
ax.set_xlabel('MJD')
ax.set_ylabel('flux')
ax.set_title('KOI-63');
```

```python
# add up-directory to Python path
import sys, os; sys.path.append(os.path.abspath('..'))
from periodogram.acf import ACF
```

```python
model = ACF(maxlag=30).fit(t,y)

periods = np.linspace(0.1, 30, 1000)
scores = model.score(periods)

plt.plot(periods, scores, label='standard');

plt.legend()
```

```python
model.period_search()
```

##ACF estimation techniques; non-evenly sampled data

To demonstrate the Scargle (1989) and Edelson-Krolik methods, we'll pull some data from AstroML. 

```python
from astroML.datasets import fetch_LINEAR_sample
data = fetch_LINEAR_sample()
```

```python
lcid = 10022663
t, y, dy = data[lcid].T
fig, ax = plt.subplots()
ax.errorbar(t, y, dy, fmt='o');
ax.invert_yaxis()
ax.set_xlabel('MJD')
ax.set_ylabel('magnitude')
ax.set_title('LINEAR object {0}'.format(lcid));
```

```python
model_scargle = ACF(maxlag=5, method='scargle', omega_max=100, n_omega=1e4, smooth=0.05).fit(t,y)
model_EK = ACF(maxlag=5, method='EK').fit(t,y)

periods = np.linspace(0.1, 5, 1000)
scores_scargle = model_scargle.score(periods)
scores_EK = model_EK.score(periods)

plt.plot(periods, scores_scargle, );
plt.plot(periods, scores_EK, label='E-K');

plt.legend()
```
