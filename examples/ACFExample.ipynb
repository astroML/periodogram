# Using the autocorrelation function to find a periodic signal


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# use seaborn for matplotlib settings
import seaborn; seaborn.set()
```

## The Data
Exmple data will be Q1-Q15 photometry of KOI-63. We're using Kepler data as the example because the straight-up ACF requires evenly-sampled data.  This data is pre-conditioned to have no gaps (gaps have been set to zero).

```python
data = np.load('koi63.npy')
t, y = (data[:,0], data[:,1])

fig, ax = plt.subplots()
ax.plot(t, y, 'o', ms=0.5);
ax.set_xlabel('MJD')
ax.set_ylabel('flux')
ax.set_title('KOI-63');
```

```python
# add up-directory to Python path
import sys, os; sys.path.append(os.path.abspath('..'))
from periodogram.acf import ACF
```

```python
model = ACF(maxlag=1000).fit(t,y)

periods = np.linspace(0.1, 20, 1000)
scores = model.score(periods)

plt.plot(periods, scores);
```
